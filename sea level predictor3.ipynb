```python
#
import pandas as pd
import numpy as np

df = pd.read_csv('epa-sea-level.csv')

df.dropna(subset=['Year', 'CSIRO Adjusted Sea Level'], inplace=True)

df['CSIRO Adjusted Sea Level'].fillna(df['CSIRO Adjusted Sea Level'].median(), inplace=True)

df['Year'] = df['Year'].astype(int)

sea_level_data = df['CSIRO Adjusted Sea Level'].values

mean_sea_level = np.mean(sea_level_data)
std_sea_level = np.std(sea_level_data)
min_sea_level = np.min(sea_level_data)
max_sea_level = np.max(sea_level_data)

print(f"Mean Sea Level: {mean_sea_level}")
print(f"Standard Deviation of Sea Level: {std_sea_level}")
print(f"Min Sea Level: {min_sea_level}")
print(f"Max Sea Level: {max_sea_level}")

df_filtered = df[df['Year'] > 2000]

correlation = df['Year'].corr(df['CSIRO Adjusted Sea Level'])

print(df_filtered.head())
print(f"Correlation between Year and Sea Level: {correlation}")

df.to_csv('cleaned_epa_sea_level.csv', index=False)
```

```python
#
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('epa-sea-level.csv')

plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['CSIRO Adjusted Sea Level'], label='Sea Level')
plt.title('CSIRO Adjusted Sea Level Over Time')
plt.xlabel('Year')
plt.ylabel('Sea Level (in meters)')
plt.grid(True)
plt.legend()
plt.show()

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.regplot(x='Year', y='CSIRO Adjusted Sea Level', data=df, scatter_kws={'s': 10}, line_kws={'color': 'red'})
plt.title('Sea Level vs Year with Regression Line')
plt.xlabel('Year')
plt.ylabel('Sea Level (in meters)')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', y='CSIRO Adjusted Sea Level', data=df)
plt.title('Sea Level Distribution by Year')
plt.xlabel('Year')
plt.ylabel('Sea Level (in meters)')
plt.show()
```
