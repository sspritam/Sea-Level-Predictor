```python
#
import numpy as np
import pandas as pd

data_frame = pd.read_csv('epa-sea-level.csv')

years = data_frame['Year'].values
sea_levels = data_frame['CSIRO Adjusted Sea Level'].values

adjusted_sea_levels = sea_levels + 0.3 * (years - 2000)

mean_sea_level = np.mean(sea_levels)
sum_sea_level = np.sum(sea_levels)

print(f"Mean Sea Level: {mean_sea_level} inches")
print(f"Sum of Sea Level Data: {sum_sea_level} inches")

recent_years = years[years > 2000]
recent_sea_levels = sea_levels[years > 2000]

combined_data = np.column_stack((years, sea_levels))

combined_adjusted_data = np.column_stack((years, adjusted_sea_levels))

vertical_combined_data = np.vstack((combined_data, combined_adjusted_data))

print("Combined Data (Years and Original Sea Levels):")
print(combined_data)

print("Combined Adjusted Data (Years and Adjusted Sea Levels):")
print(combined_adjusted_data)

print("Vertical Combined Data (Original and Adjusted):")
print(vertical_combined_data)

high_sea_levels_years = years[sea_levels > 7]
high_sea_levels_values = sea_levels[sea_levels > 7]

print("Years with Sea Levels Greater Than 7 Inches:")
print(high_sea_levels_years)
print(high_sea_levels_values)
```

```python
#
import numpy as np
import pandas as pd

data_frame = pd.read_csv('epa-sea-level.csv')

years = data_frame['Year'].values
sea_levels = data_frame['CSIRO Adjusted Sea Level'].values

year_2000_sea_level = sea_levels[years == 2000][0]
sea_level_change = sea_levels - year_2000_sea_level

print("Sea Level Change (Relative to Year 2000):")
print(sea_level_change)

scaling_factor = 0.2
adjusted_sea_levels = sea_levels + scaling_factor * (years - 2000)

print("Adjusted Sea Levels:")
print(adjusted_sea_levels)

years_above_7 = years[sea_levels > 7]
sea_levels_above_7 = sea_levels[sea_levels > 7]

print("Years with Sea Levels Greater Than 7 Inches:")
print(years_above_7)
print(sea_levels_above_7)

combined_data = np.column_stack((years, adjusted_sea_levels))

print("Combined Data (Years and Adjusted Sea Levels):")
print(combined_data)
```
